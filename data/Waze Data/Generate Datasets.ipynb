# Neighborhood

```python
def makeQuery(f, chunk=0, chunk_size=100):
    df = pd.read_csv(f)
    ring = list({r for h in df.h3id for r in h3.k_ring(h, 3)} - set(df.h3id))
    st = chunk*chunk_size
    end = (chunk+1)*chunk_size
    c = ring[st:end]
    print('--', f, f'{st}-{end}' ,'/' ,len(ring))
    print(many_h3_to_query(c))
```
