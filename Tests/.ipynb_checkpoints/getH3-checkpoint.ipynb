### In the codes below, I get the H3 id within a circumference.
### The center of the circumference is a station.

```python
import h3
import folium
import pandas as pd
import numpy as np
```

```python
def findpoints(lat,lon,rad):
    
    N = 360 

    # list of latitudes
    latitude = []
    # list of longitudes
    longitude = []
    
    for r in range(1,rad*10):
        
        radius = r/10
        
        for k in range(N):
            
            # angle in radians
            angle = np.pi*2*k/N
            
            dx = radius*np.cos(angle)
            dy = radius*np.sin(angle)
            
            latitude.append(lat + (180/np.pi)*(dy/6371)) #Earth Radius
            longitude.append(lon + (180/np.pi)*(dx/6371)/np.cos(lon*np.pi/180)) #Earth Radius
        
    return latitude,longitude
```

```python
# load all latitude and longitude for stations
df=pd.read_csv('latlongAll.csv')
df
```

```python
# Get latitude and longitude within a circumference

# centers of circumferences 
centers=[[-23.553291,-46.673462],
[-23.616203,-46.662945],
[-23.587416,-46.657634],
[-23.680984,-46.675839],
[-23.501934,-46.420393],
[-23.518655,-46.74119],
[-23.54408,-46.627352],
[-23.458247,-46.76686],
[-23.56119,-46.701544]]


# Get latitudes and longitudes
dataLat=[]
dataLong=[]
radius=2   # 2km

for i in range(np.shape(centers)[0]):
    temp1,temp2 =findpoints(centers[i][0],centers[i][1],2)
    dataLat.extend(temp1)
    dataLong.extend(temp2)
```

```python
# Get H3 indexes

h3Id=[]

for i in range(np.size(dataLat)):
    h3Id.append(h3.geo_to_h3(dataLat[i], dataLong[i], 9))
```

```python
# create dataframe
df=pd.DataFrame({'h3id':h3Id,'Latitude':dataLat,'Longitude':dataLong})
# remove duplicates
df.drop_duplicates(subset=['h3id'],inplace=True)
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
m = folium.Map(location=[df.iloc[0]['Latitude'], df.iloc[0]['Longitude']],zoom_start=12)
for i in range(df.shape[0]):
    # add hexagon to folium
    folium.Polygon(h3.h3_to_geo_boundary(df.iloc[i]['h3id']),tooltip=i,color= "black", fill_color= "blue", fill_opacity= 0.5).add_to(m)
```
